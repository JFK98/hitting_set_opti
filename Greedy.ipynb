```python
#SIN K
import time

def validar_entrada(U, S):
    U = set(U)
    for subconjunto in S:
        for x in subconjunto:
            if x not in U:
                print("Error:", x, "no pertenece a U")
                return False
    return True

def greedy(U, S):
    restantes = S.copy()
    H = []

    while len(restantes) > 0:
        mejor_elemento = None
        max_cobertura = 0

        for elemento in U:
            cobertura = 0

            for subconjunto in restantes:
                if elemento in subconjunto:
                    cobertura += 1

            if cobertura > max_cobertura:
                max_cobertura = cobertura
                mejor_elemento = elemento

        if mejor_elemento is None:
            return None

        H.append(mejor_elemento)

        nuevos = []
        for subconjunto in restantes:
            if mejor_elemento not in subconjunto:
                nuevos.append(subconjunto)

        restantes = nuevos

    return H

def ejecutar(U, S, nombre):
    print("\n---", nombre, "---")

    if validar_entrada(U, S):
        inicio = time.perf_counter()
        solucion = greedy(U, S)
        fin = time.perf_counter()

        tiempo_ms = (fin - inicio) * 1000

        print("Solución:", solucion)
        print("Tamaño:", len(solucion))
        print("Tiempo:", round(tiempo_ms, 4), "ms")
    else:
        print("Datos inválidos")

U1 = [1, 2, 3, 4, 5]
S1 = [{1, 2}, {2, 3}, {3, 4}, {4, 5}]

U2 = [1, 2, 3, 4, 5, 6, 7, 8]
S2 = [{1, 2, 3}, {2, 4, 5}, {3, 6}, {5, 7}, {6, 8}, {1, 7}, {2, 8}]

U3 = list(range(1, 13))
S3 = [
    {1, 2, 3, 4}, {2, 5, 6}, {3, 7, 8}, {4, 9},
    {5, 10}, {6, 11}, {7, 9}, {8, 10},
    {1, 11}, {2, 7}, {3, 5}
]

ejecutar(U1, S1, "Caso 1")
ejecutar(U2, S2, "Caso 2")
ejecutar(U3, S3, "Caso 3")
```

```python
#CON K
import time

def validar_entrada(U, S):
    U = set(U)
    for subconjunto in S:
        for x in subconjunto:
            if x not in U:
                print("Error:", x, "no pertenece a U")
                return False
    return True

def greedy_k(U, S, k):
    restantes = S.copy()
    H = []

    while len(restantes) > 0:
        mejor_elemento = None
        max_cobertura = 0

        for elemento in U:
            cobertura = 0

            for subconjunto in restantes:
                if elemento in subconjunto:
                    cobertura += 1

            if cobertura > max_cobertura:
                max_cobertura = cobertura
                mejor_elemento = elemento

        if mejor_elemento is None:
            return None

        H.append(mejor_elemento)

        if len(H) > k:
            return None

        nuevos = []
        for subconjunto in restantes:
            if mejor_elemento not in subconjunto:
                nuevos.append(subconjunto)

        restantes = nuevos

    return H

def ejecutar(U, S, k, nombre):
    print("\n---", nombre, "---")
    print("k:", k)

    if validar_entrada(U, S):
        inicio = time.perf_counter()
        solucion = greedy_k(U, S, k)
        fin = time.perf_counter()

        tiempo_ms = (fin - inicio) * 1000

        if solucion is not None:
            print("Solución:", solucion)
            print("Tamaño:", len(solucion))
        else:
            print("No existe solución con k =", k)

        print("Tiempo:", round(tiempo_ms, 4), "ms")
    else:
        print("Datos inválidos")

U1 = [1, 2, 3, 4, 5]
S1 = [{1, 2}, {2, 3}, {3, 4}, {4, 5}]
k1 = 2

U2 = [1, 2, 3, 4, 5, 6, 7, 8]
S2 = [{1, 2, 3}, {2, 4, 5}, {3, 6}, {5, 7}, {6, 8}, {1, 7}, {2, 8}]
k2 = 3

U3 = list(range(1, 13))
S3 = [
    {1, 2, 3, 4}, {2, 5, 6}, {3, 7, 8}, {4, 9},
    {5, 10}, {6, 11}, {7, 9}, {8, 10},
    {1, 11}, {2, 7}, {3, 5}
]
k3 = 4

ejecutar(U1, S1, k1, "Caso 1")
ejecutar(U2, S2, k2, "Caso 2")
ejecutar(U3, S3, k3, "Caso 3")
```
