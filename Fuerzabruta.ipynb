```python
#SIN K
import time

def validar_entrada(U, S):
    U = set(U)
    for subconjunto in S:
        for x in subconjunto:
            if x not in U:
                print("Error:", x, "no pertenece a U")
                return False
    return True

def es_hitting_set(H, S):
    for subconjunto in S:
        cumple = False
        for x in H:
            if x in subconjunto:
                cumple = True
                break
        if not cumple:
            return False
    return True

def fuerza_bruta(U, S):
    U = list(U)
    n = len(U)
    mejor = None

    for i in range(1, 2 ** n):
        H = []

        for j in range(n):
            if (i >> j) & 1:
                H.append(U[j])

        if es_hitting_set(H, S):
            if mejor is None or len(H) < len(mejor):
                mejor = H

    return mejor

def ejecutar(U, S, nombre):
    print("\n---", nombre, "---")

    if validar_entrada(U, S):
        inicio = time.perf_counter()
        solucion = fuerza_bruta(U, S)
        fin = time.perf_counter()

        tiempo_ms = (fin - inicio) * 1000

        print("Solución:", solucion)
        print("Tamaño:", len(solucion))
        print("Tiempo:", round(tiempo_ms, 4), "ms")
    else:
        print("Datos inválidos")

U1 = [1, 2, 3, 4, 5]
S1 = [{1, 2}, {2, 3}, {3, 4}, {4, 5}]

U2 = [1, 2, 3, 4, 5, 6, 7, 8]
S2 = [{1, 2, 3}, {2, 4, 5}, {3, 6}, {5, 7}, {6, 8}, {1, 7}, {2, 8}]

U3 = list(range(1, 13))
S3 = [
    {1, 2, 3, 4}, {2, 5, 6}, {3, 7, 8}, {4, 9},
    {5, 10}, {6, 11}, {7, 9}, {8, 10},
    {1, 11}, {2, 7}, {3, 5}
]

ejecutar(U1, S1, "Caso 1")
ejecutar(U2, S2, "Caso 2")
ejecutar(U3, S3, "Caso 3")
```

```python
#CON K
import time

def validar_entrada(U, S):
    U = set(U)
    for subconjunto in S:
        for x in subconjunto:
            if x not in U:
                print("Error:", x, "no pertenece a U")
                return False
    return True

def es_hitting_set(H, S):
    for subconjunto in S:
        cumple = False
        for x in H:
            if x in subconjunto:
                cumple = True
                break
        if not cumple:
            return False
    return True

def fuerza_bruta_k(U, S, k):
    U = list(U)
    n = len(U)
    mejor = None

    for i in range(1, 2 ** n):
        H = []

        for j in range(n):
            if (i >> j) & 1:
                H.append(U[j])

        if len(H) <= k and es_hitting_set(H, S):
            if mejor is None or len(H) < len(mejor):
                mejor = H

    return mejor

def ejecutar(U, S, k, nombre):
    print("\n---", nombre, "---")
    print("k:", k)

    if validar_entrada(U, S):
        inicio = time.perf_counter()
        solucion = fuerza_bruta_k(U, S, k)
        fin = time.perf_counter()

        tiempo_ms = (fin - inicio) * 1000

        if solucion is not None:
            print("Solución:", solucion)
            print("Tamaño:", len(solucion))
        else:
            print("No existe solución con k =", k)

        print("Tiempo:", round(tiempo_ms, 4), "ms")
    else:
        print("Datos inválidos")

U1 = [1, 2, 3, 4, 5]
S1 = [{1, 2}, {2, 3}, {3, 4}, {4, 5}]
k1 = 2

U2 = [1, 2, 3, 4, 5, 6, 7, 8]
S2 = [{1, 2, 3}, {2, 4, 5}, {3, 6}, {5, 7}, {6, 8}, {1, 7}, {2, 8}]
k2 = 3

U3 = list(range(1, 13))
S3 = [
    {1, 2, 3, 4}, {2, 5, 6}, {3, 7, 8}, {4, 9},
    {5, 10}, {6, 11}, {7, 9}, {8, 10},
    {1, 11}, {2, 7}, {3, 5}
]
k3 = 4

ejecutar(U1, S1, k1, "Caso 1")
ejecutar(U2, S2, k2, "Caso 2")
ejecutar(U3, S3, k3, "Caso 3")
```
